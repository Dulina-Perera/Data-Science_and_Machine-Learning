## Linear Regression

```python
import matplotlib.pyplot as plt  # type: ignore
import numpy as np  # type: ignore

from sklearn.linear_model import LinearRegression  # type: ignore
from sklearn.preprocessing import PolynomialFeatures  # type: ignore

np.random.seed(42)
```

```python
X: np.ndarray = np.random.rand(100, 1)
y: np.ndarray = 4 + 5 * X + np.random.normal(0, 0.5, (100, 1))
```

```python
reg: LinearRegression = LinearRegression()
reg.fit(X, y)

print(f"Intercept: {reg.intercept_} \nCoefficient: {reg.coef_}")
```

```python
X_ticks: np.ndarray = np.linspace(0, 1, 100).reshape(-1, 1)
y_preds: np.ndarray = reg.predict(X_ticks)

plt.scatter(X, y)
plt.plot(X_ticks, y_preds, color="red")
plt.show()
```

## Polynomial Regression

```python
X = 4 * np.random.rand(200, 1) - 2
y = 4 - 7 * X + 19 * X**2 + 7 * X**3 + np.random.normal(0, 5, (200, 1))
```

```python
reg = LinearRegression()
reg.fit(X, y)

print(f"Intercept: {reg.intercept_} \nCoefficient: {reg.coef_}")
```

```python
poly_features: PolynomialFeatures = PolynomialFeatures(degree=3, include_bias=False)
X_poly: np.ndarray = poly_features.fit_transform(X)

reg_poly: LinearRegression = LinearRegression()
reg_poly.fit(X_poly, y)

print(f"Intercept: {reg_poly.intercept_} \nCoefficient: {reg_poly.coef_}")
```

```python
X_ticks: np.ndarray = np.linspace(-2, 2, 200).reshape(-1, 1)
X_ticks_poly: np.ndarray = poly_features.fit_transform(X_ticks)

y_preds: np.ndarray = reg.predict(X_ticks)
y_preds_poly: np.ndarray = reg_poly.predict(X_ticks_poly)

plt.scatter(X, y)
plt.plot(X_ticks, y_preds, color="red")
plt.plot(X_ticks, y_preds_poly, color="green")
plt.show()
```
